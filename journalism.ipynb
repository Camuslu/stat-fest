```python
sw = set(stopwords.words('english'))
punctuation = set(string.punctuation)
all_names = set([name.lower() for name in names.words()])

def isStopWord(word):
	return (word in sw or word in punctuation) or not word.isalpha() or word in all_names
```

```python
row_disease=defaultdict(str)
```

```python
counter = 0
for row in range(ny.shape[0]):
    argmax=defaultdict(int)
    string=str(ny['Name_of_Study'][row])
    if string in row_disease.keys():
        ny['disease'][row]=row_disease[string]
    else:
        for flag in string.split(' '):
            if not isStopWord(flag.lower()) and len(flag) > 1:
                for i in range(len(allD)):
                    if flag.lower() in map(lambda x:x.lower(), allD[i].keys()):
                        argmax[alld[i]]+=1
        mle=sorted(argmax,key=argmax.get,reverse=True)
        row_disease[string]=mle
        ny['disease'][row]=mle
    counter+=1
    if counter%1000==0:
        print counter
        
```
